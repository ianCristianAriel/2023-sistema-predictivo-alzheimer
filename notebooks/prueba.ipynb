```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/ADNI_Training_Q3_APOE_CollectionADNI1Complete 1Yr 1.5T_July22.2014.csv')
```

```python
df.sample(4, random_state=3)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que 'df' es tu DataFrame

# Itera a través de cada columna y crea un gráfico de barras para mostrar la distribución
for column in df.columns:
    # Filtra columnas de tipo 'object' (cadenas u otras categorías)
    value_counts = df[column].value_counts()
    plt.figure(figsize=(10, 6))  # Tamaño del gráfico
    plt.bar(value_counts.index, value_counts.values)
    plt.xlabel(column)  # Etiqueta del eje x con el nombre de la columna
    plt.ylabel('Count')  # Etiqueta del eje y
    plt.title(f'Distribution of {column}')  # Título del gráfico
    plt.xticks(rotation=45)  # Opcional: Rota las etiquetas del eje x para mejor visibilidad
    plt.show()
```
